```python
def parse_rda(filename):
    header = {}

    with open(filename, 'rb') as file_binary:
        for line_binary in file_binary:
            line = (str(line_binary)
                .replace('b\'', '')
                .replace('b"', '')
                .replace('\\r\\n\'', '')
                .replace('\\r\\n"', ''))
            if ('Begin of header' in line):
                continue
            if ('End of header' in line):
                break

            key, val = line.split(': ')

            header[key] = val
        
    return header
```

```python
import glob
import re
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(14,10))
ax = fig.add_subplot(111, projection='3d')

scans = glob.glob('/PHShome/hj095/mnt/spectro_data/Study_2HG/NicolettiG*')

def roi(name):
    print(name)
    rdas = []
    for scan in scans:
        rdas.extend(
            [x for x in glob.glob('{}/*{}.rda'.format(scan, name)) if (
                ('_30_' not in x)
                and ('_40_' not in x)
                and ('_wref' not in x)
            )]
        )

    print(*rdas, sep = "\n")
    coords = ([], [], [])

    print('date     sag        cor       tra')
    for rda in rdas:
        headers = parse_rda(rda)
        if (headers['TE'] == '97.000000'):
            print(headers['StudyDate'], headers['VOIPositionSag'], headers['VOIPositionCor'], headers['VOIPositionTra'])
            coords[0].append(float(headers['VOIPositionSag']))
            coords[1].append(float(headers['VOIPositionCor']))
            coords[2].append(float(headers['VOIPositionTra']))

    ax.scatter(coords[0], coords[1], coords[2], label=name, marker='D', size=5)
    print()
    
roi('leftsup')
roi('leftsup_CL')
roi('postlat')
roi('postmed')

ax.set_xlabel('Sag')
ax.set_ylabel('Cor')
ax.set_zlabel('Tra')

ax.legend()

fig.suptitle('NicolettiG')

plt.show()
```
